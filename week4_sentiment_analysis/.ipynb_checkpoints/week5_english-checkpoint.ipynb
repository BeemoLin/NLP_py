```python
def re_rule(df):
    for i in range(len(df)):
        #text = re.sub('\n', '', df.iloc[i,9]) #remove /n
        #df.iloc[i,9] = re.sub('[A-Za-z0-9_|!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]+', '', text) # remove alphabet, number and punctuation
        df.iloc[i,9] = re.sub('[^\u4e00-\u9fa5]+', '', str(df.iloc[i,9]))  # only remain Chinese words
```

### (2) Word Segmentation
   - pip install jieba

```python
import jieba
import jieba.analyse
```

```python
# set user dictionary
jieba.load_userdict('./dict/user_dict.txt')
# add stop words
#jieba.analyse.set_stop_words('./dict/stop_words.txt') #only work on jieba.analyse.extract_tags
with open('./dict/stop_words.txt',encoding="utf-8") as f:
    stopWords = [line.strip() for line in f.readlines()]
```

```python
# Word Segmentation function
def getToken(df):
    df['word'] = 0
    for i in range(len(df)):
        seg_list = jieba.lcut(str(df.iloc[i,9]))
        seg_list = [w for w in seg_list if w not in stopWords]
        df['word'][i] = seg_list
    df = df.explode('word')
    
    return df


# def getToken(df, text = 'text'):
#     df = (df
#           .assign(word=df.iloc[:,9].apply(jieba.lcut).apply(lambda seg_list: [w for w in seg_list if w not in stopWords]))
#           .explode('word'))
#     return df
```

```python
MetaData = getToken(MetaData)
MetaData.head(3)
```

```python
Reviews = getToken(Reviews)
Reviews.head(3)
```

```python
# Combine MetaData and Reviews tables
data = pd.concat([MetaData.loc[:,["artDate","artUrl","word"]],Reviews.loc[:,["artDate","artUrl","word"]]],axis=0)
data.head()
```

```python
# convert date string to datetime format
from datetime import datetime
data['artDate'] = pd.to_datetime(data['artDate'], format="%Y/%m/%d")
```

```python
# count words frequence everyday
word_count = data.groupby(['artDate','word'],as_index=False).size()
word_count.sort_values(by=['size'], ascending=False,inplace=True)  #sort
word_count = word_count[word_count['size']>3]  # filter words frequence above 3
word_count
```

## 2. LIWC Dictionary  
- Linguistic Inquiry and Word Counts，publish by Pennebaker in 2001  
- can divided into positive and negative

```python
filepath = ['negative','positive']
d_liwc = []
for file in filepath:
    filename = f"./dict/liwc/{file}.txt"
    with open(filename, 'r', encoding='UTF-8-sig') as f:
        d_liwc.append([f.read().split(","),file])
```

```python
df_liwc = pd.DataFrame(d_liwc,columns = ['word','sentiments']).explode('word',ignore_index = True)
df_liwc
```

## 4. Draw WordCloud
- draw word cloud to see what are the discussion on article
- pip install wordcloud

### 2021-03-19 WordCloud
The wordcloud appear words "waste", "greed for cheap", "humiliation" and "beggar".  
It must be because netizens criticized the unwise behavior of changing the name.  
Some people are also criticized for who eat free sushi and causing food waste.

```python
from wordcloud import WordCloud 
```

```python
def Wc_dict(df,Date,countB,countS):
    wc_df = (df[(df['artDate']==Date) & (df['word']!="鮭魚")].loc[:, ['word', 'size']]  #set the date，remove word "鮭魚"，keep word, size column
             .groupby('word', as_index=False)
             .agg({'size': 'sum'})  #count the sum of size after groupby
             .query('size < @countB')  #filter size < count
             .query('size > @countS')) 
    wc_dict = dict(zip(wc_df['word'], wc_df['size'])) #set as dictionary
    
    return wc_dict
    
```

```python
dict_0319 = Wc_dict(word_count,"2021-03-19",500,50)

font = './data/SourceHanSansTW-Regular.otf' #font path for Chinese font
wordcloud = WordCloud(background_color='white', font_path=font, colormap='viridis')
wordcloud.generate_from_frequencies(dict_0319)
plt.imshow(wordcloud)
plt.axis("off") 
plt.show()
```

### 2021-03-17 WordCloud
There are less negative words and more positive words such as "funny", "young people", "marketing", "advertising" in 3/17.  
It is speculated that there are fewer negative articles at this time.  
Netizens still hold a more humorous attitude towards this topic.  
They think the project is funny and young people are too impulsive.

```python
dict_0317 = Wc_dict(word_count,"2021-03-17",500,50)

font = './data/SourceHanSansTW-Regular.otf' 
wordcloud = WordCloud(background_color='white', font_path=font, colormap='viridis')
wordcloud.generate_from_frequencies(dict_0317)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## 6. Categorize positive and negative articles
- positive and negative articles separately
- in the following, we count the sentiment score per article to determine whether the article is positive or negative.

```python
article_type = (data.merge(df_liwc, on='word')
                .groupby(['artUrl', 'sentiments'], as_index=False)
                .size()
                .pivot_table(index='artUrl', columns='sentiments', values='size', fill_value=0)
                .reset_index(['artUrl'])
               )
article_type.head()
```

```python
article_type['type'] = article_type.apply(lambda x: 'positive' if x['positive'] > x['negative'] else 'negative',axis=1)
article_type.groupby('type', as_index=False).size()
```

### Bar chart of positive and negative articles

```python
article_type_date = article_type.merge(MetaData.loc[:,["artUrl", "artDate"]], how = 'left', on = 'artUrl')
article_type_date = article_type_date.groupby(['artDate','type'], as_index=False).size()
#article_type_date
```

```python
ap = article_type_date[article_type_date["type"]=="positive"]
an = article_type_date[article_type_date["type"]=="negative"]

plt.figure(figsize=(5,3))   
plt.style.use("ggplot")    
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.title('Bar chart of positive and negative articles')
plt.xlabel('artDate')
plt.bar(ap['artDate'],ap['size'],align="edge",width=0.35,label="positive")
plt.bar(an['artDate'],an['size'],align="edge",width=-0.35,label="negative")
plt.legend()
plt.xticks(rotation= 45, fontsize=9)
plt.xlabel('artDate')
plt.show()
```

```python
p_sum = positive_article[positive_article["sentiments"]=="positive"].head(15).sort_values(by=['size'])
n_sum = positive_article[positive_article["sentiments"]=="negative"].head(15).sort_values(by=['size'])

plt.figure(figsize=(8,5))   
plt.style.use("ggplot")   
plt.rcParams['font.sans-serif']=['SimHei'] 

plt.subplot(121)
plt.title('Positive')
plt.xlabel('count')
plt.barh(p_sum['word'],p_sum['size'])

plt.subplot(122)
plt.title('Negative')
plt.xlabel('count')
plt.barh(n_sum['word'],n_sum['size'],color="darkblue")


plt.show()
```

```python
# Word Segmentation
Dcard['word'] = 0
for i in range(len(Dcard)):
    seg_list = jieba.lcut(str(Dcard.loc[i,"sentence"]))
    seg_list = [w for w in seg_list if w not in stopWords]
    Dcard['word'][i] = seg_list
DToken = Dcard.explode('word')
```

```python
DToken["source"] = "Dcard"
DToken['artDate'] = pd.to_datetime(DToken['artDate'], format="%Y/%m/%d")
PToken = data
PToken["source"] = "PTT"
```

```python
Dcard_senti_sum = (DToken.loc[:,["artDate","artUrl","word","source"]]
                   .merge(df_liwc, on='word')
                   .groupby(['artDate', 'sentiments'], as_index=False)
                   .size()
                   .query('artDate >= "2021-03-15"'))
```

```python
PTT_senti_sum = (PToken
                 .merge(df_liwc, on='word')
                 .groupby(['artDate', 'sentiments'], as_index=False)
                 .size()
                 .query('artDate >= "2021-03-15"'))
```

### compare ptt and dcard
